## 関数をつかえるようになる

```python
# 組み込み関数はimport文なしで使える
# 2の3乗を計算
# 丸括弧（）の中に引数（ひきすう）として関数にデータを渡す

pow(2, 3)
```

```python
# 組み込み関数以外の関数はimportで読み込んでから
import math

# 4!を計算
math.factorial(4)
```

```python
#　ここで「もの」と書いた部分は、オブジェクトと表現されることがある
#　Pythonはオブジェクト指向言語の一種なので、すべてがオブジェクト
# 実は関数自体もオブジェクト
type(math.factorial)
```

```python
#　いろいろな関数がある
# 乱数を使ってみる
# 公式ドキュメントを読む習慣を付けるのは良いです（が、最初はすこし戸惑うかも）
# https://docs.python.org/3/library/random.html

import random
```

```python
# 0 <= X < 1.0の乱数を発生させる
# 引数のない関数
random.random()
```

```python
# 何度か実行すると違う結果が返る
for _ in range(10):
    print(random.random())
```

```python
# コンピュータのなかの乱数は疑似乱数列
# seedを固定すると同じ数列が返ってくる
# Random Forestsなど内部で乱数を利用する機械学習アルゴリズムは多く、seedを固定すると同じ結果を得られる（と期待される）
random.seed(1)

for _ in range(10):
    print(random.random())
```

## オブジェクト指向に慣れる

おおくのプログラミング言語があり、それぞれに流派のようなものがある。これはプログラミングのパラダイムと呼ばれる。パラダイムには、手続き型、関数型、オブジェクト指向、宣言型などいくつかあって、複数のパラダイムの影響を受けている言語も多い。Pythonは手続き型に基盤を置いたオブジェクト指向言語であり、関数型言語の影響も受けている。全部理解していると言語オタクになってしまうので、まずはオブジェクト指向に慣れることが重要。

```python
# 再び日付を表現するオブジェクト
#　初期化メソッドやコンストラクタと呼ばれる特殊な関数を使って、オブジェクトを作っている
# オブジェクトの具体的な実態をインスタンスと呼び、オブジェクトの設計図をクラスと言うことも（このあたりの用語はそれほど神経質にならなくてもOK）

today = dt.datetime(2023, 9, 22)
```

```python
#　オブジェクトは、データと関数（メソッドと呼ぶ）の塊
#　組み込み関数dirで何をもっているのか、確認できる
# 説明は、公式ドキュメントを参照
# https://docs.python.org/3/library/datetime.html

dir(today)
```

```python
# 年月日のデータは、属性（アトリビュート）と呼ばれる変数で参照
today.year
```

```python
today.month
```

```python
today.day
```

```python
# オブジェクト指向のすごいところは、内部に保持したデータから、複雑な計算の仕組みを隠蔽してくれるところ
# たとえば、weekday()メソッドは、その日が何曜日か計算してくれる（月曜日が0）

today.weekday()
```

```python
#　1/1/1から何日経ったか？（そんなこと知りたいかは別として）
# ちなみに、2023＊365＝738395

today.toordinal()
```

```python
# 画像を使った例
# 外部ライブラリのPillow
# https://pillow.readthedocs.io/en/stable/index.html

from PIL import Image

# PNGファイルのファイル名を引数にして、Imageオブジェクトを作る
tiger = Image.open("tiger.png")
```

```python
# メソッドを呼び出すとシステムのデフォルトアプリケーションが起動
tiger.show()
```

```python
# 白黒に変換
# convertメソッドに引数”L"で指示を出している

tiger.convert('L')
```

```python
# 白黒に変換した画像をjpegで保存
# メソッドチェインと言うことも

tiger.convert('L').save('tiger.jpeg')
```
